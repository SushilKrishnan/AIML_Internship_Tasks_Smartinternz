Outlier detection

```python
sns.boxplot(df.bmi)
```

```python
#Outlier replacement with median
q1 = df.bmi.quantile(0.25)
q3 = df.bmi.quantile(0.75)
IQR = q3 - q1
upper_limit = q3+1.5*IQR    #Range after which we have outlier in boxplot
lower_limit = q1-1.5*IQR    #Range below which we have outlier in boxplot
median = df['bmi'].median()
df['bmi'] = np.where(df['bmi']>upper_limit, median, df['bmi'])
```

```python
sns.boxplot(df.bmi)
```

Ridge and Lasso regression

```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/50_Startups.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.State = le.fit_transform(df.State)
```

```python
x = df.drop(columns = ['Profit'], axis = 1)
y = df.Profit
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
x_sc = pd.DataFrame(scale.fit_transform(x), columns = x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_tr, x_te, y_tr, y_te = train_test_split(x_sc, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
r=Ridge()
l=Lasso()
r.fit(x_tr,y_tr)
l.fit(x_tr,y_tr)
pred_r = r.predict(x_te)
pred_l = l.predict(x_te)
profit = pd.DataFrame({'Actual Profit':y_te, 'Ridge Prediction': pred_r, 'Lasso Prediction': pred_l})
profit
```

```python
from sklearn import metrics
print("Ridge r2: "+str(metrics.r2_score(y_te,pred_r)))
print("Lasso r2: "+str(metrics.r2_score(y_te,pred_l)))
print("Ridge MSE: "+str(metrics.mean_squared_error(y_te,pred_r)))
print("Lasso MSE: "+str(metrics.mean_squared_error(y_te,pred_l)))
print("Ridge RMSE: "+str(np.sqrt(metrics.mean_squared_error(y_te,pred_r))))
print("Lasso RMSE: "+str(np.sqrt(metrics.mean_squared_error(y_te,pred_l))))
```

```python
import pandas as pd
import numpy as np
import nltk
import regex as re

nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

Text preprocessing

```python
a = "Hi all!!!, I'm Sushil Krishnan. Welcome to Vellore."
a
```

```python
#step 1 - remove special characters using regex
reg = re.sub('[^a-zA-Z]',' ',a)
reg
```

```python
#step 2 - convert entire text into upper or lower case
lw = reg.lower()
lw
```

```python
#step 3 - split the input
wd = lw.split()
wd
```

```python
#step 4 - removing stopwords
print("stopwords of english: "+str(stopwords.words('english')))
sw = [w for w in wd if w not in set(stopwords.words("english"))]
print(sw)
```

```python
#step 5 - stemming on data (Shortform of original value ex:pay vs paid)
ps = PorterStemmer()
p = [ps.stem(w) for w in sw]
p
```

```python
#step 6 - Join the words
st = ' '.join(p)
st
```

```python
#step 7 - convert text into number
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
cv.fit_transform([st]).toarray()
```

With dataset

```python
df = pd.read_csv("/content/Restaurant_Reviews.tsv",delimiter='\t')
df
```

```python
data = []
for i in range(0,1000):
  review = df['Review'][i]    #read text review
  review = re.sub('[^a-zA-Z]',' ',review)     #remove spl char
  review = review.lower()         #convert to lower case
  review = review.split()         #splitting into words
  review = [ps.stem(w) for w in review if w not in set(stopwords.words('english'))]   #remove stop words and performing stemming
  review = ' '.join(review)   #joining the words
  data.append(review)     #add to list for training
```

```python
# conver to numerical
x = cv.fit_transform(data).toarray()
x
```

```python
# taking liked from dataset
y = df['Liked'].values
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.3,random_state=42)
```

Model building - Naive Bayes

```python
#Model building
from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
model.fit(x_tr,y_tr)
```

```python
y_pred = model.predict(x_te)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_te,y_pred)
```

```python
pd.crosstab(y_te,y_pred)
```
